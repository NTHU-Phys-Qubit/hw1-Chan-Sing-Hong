```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=\sqrt{x^2_f+y^2_f}$ for an ensemble of trajectories.

Set each random walk runs 1000 steps and run 1000 times random walk for statistics.

```python
n_step = 60000
n_run = 1000
```

(a) Let $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.

```python
def RandomWalk_2D_1(n_step, n_run):
    x_f = np.empty(0)
    y_f = np.empty(0)
    for i in range(n_run):
        dx = np.random.random(n_step) - 0.5
        dy = np.random.random(n_step) - 0.5
        x = np.zeros(n_step + 1)
        y = np.zeros(n_step + 1)
        for j in range(n_step):
            x[j + 1] = x[j] + dx[j]
            y[j + 1] = y[j] + dy[j]
        x_f = np.append(x_f, x[-1])
        y_f = np.append(y_f, y[-1])
    return x, y, x_f, y_f
```

Check the position distribution after one step.

```python
_, _, x_f, y_f = RandomWalk_2D_1(1, n_run)
plt.figure(figsize=(5,5))
plt.plot(x_f, y_f, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('distribution after one step')
```

* Plot a single trajectory. 

```python
x, y, _, _ = RandomWalk_2D_1(n_step, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('single trajectory')
```

* Plot the distribution of the end point for an ensemble of trajectories.

```python
_, _, x_f, y_f = RandomWalk_2D_1(n_step, n_run)
plt.figure(figsize=(5,5))
plt.plot(x_f, y_f, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('distribution of end point')
```

* Plot the histogram of the $r=\sqrt{x^2_f+y^2_f}$ for an ensemble of trajectories.

```python
r = np.sqrt(x_f**2 + y_f**2)
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(r, bins = 50, density = True)
ax.set_xlabel('r')
ax.set_ylabel('probablity density')

sigma = np.average(r)*np.sqrt(2/np.pi)
x = np.linspace(0, 300)
ax.plot(x, (x/sigma**2)*np.exp(-x**2/(2*sigma**2)), linewidth=2, color='r')
```

(b) $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.

```python
def RandomWalk_2D_2(n_step, n_run):
    x_f = np.empty(0)
    y_f = np.empty(0)
    for i in range(n_run):
        dx = (-1)**np.random.randint(2, size=n_step)
        dy = (-1)**np.random.randint(2, size=n_step)
        x = np.zeros(n_step + 1)
        y = np.zeros(n_step + 1)
        for j in range(n_step):
            x[j + 1] = x[j] + dx[j]
            y[j + 1] = y[j] + dy[j]
        x_f = np.append(x_f, x[-1])
        y_f = np.append(y_f, y[-1])
    return x, y, x_f, y_f
```

Check the position distribution after one step.

```python
_, _, x_f, y_f = RandomWalk_2D_2(1, n_run)
plt.figure(figsize=(5,5))
plt.plot(x_f, y_f, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('distribution after one step')
```

* Plot a single trajectory. 

```python
x, y, _, _ = RandomWalk_2D_2(n_step, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('single trajectory')
```

* Plot the distribution of the end point for an ensemble of trajectories.

```python
_, _, x_f, y_f = RandomWalk_2D_2(n_step, n_run)
plt.figure(figsize=(5,5))
plt.plot(x_f, y_f, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('distribution of end point')
```

* Plot the histogram of the $r=\sqrt{x^2_f+y^2_f}$ for an ensemble of trajectories.

```python
r = np.sqrt(x_f**2 + y_f**2)
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(r, bins = 50, density = True)
ax.set_xlabel('r')
ax.set_ylabel('probablity density')

sigma = np.average(r)*np.sqrt(2/np.pi)
x = np.linspace(0, 800)
ax.plot(x, (x/sigma**2)*np.exp(-x**2/(2*sigma**2)), linewidth=2, color='r')
```

(c) $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

```python
def RandomWalk_2D_3(n_step, n_run):
    x_f = np.empty(0)
    y_f = np.empty(0)
    for i in range(n_run):
        theta = np.random.random(n_step)*2*np.pi
        dx = np.cos(theta)
        dy = np.sin(theta)
        x = np.zeros(n_step + 1)
        y = np.zeros(n_step + 1)
        for j in range(n_step):
            x[j + 1] = x[j] + dx[j]
            y[j + 1] = y[j] + dy[j]
        x_f = np.append(x_f, x[-1])
        y_f = np.append(y_f, y[-1])
    return x, y, x_f, y_f
```

```python
x, y, _, _ = RandomWalk_2D_3(n_step, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('single trajectory')
```

```python
_, _, x_f, y_f = RandomWalk_2D_3(n_step, n_run)
plt.figure(figsize=(5,5))
plt.plot(x_f, y_f, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('distribution of end point')
```
